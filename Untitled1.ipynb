```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r'C:\Users\Admin\Desktop\mtcars.csv')

plt.figure(figsize=[4,7])
sns.boxplot(df['mpg'])
#plt.title('boxplot')
plt.show()

plt.figure(figsize=[4,4])
plt.pie(df['am'].value_counts(),autopct='%1.1f%%',startangle=45)
#plt.title('pie chart')
plt.show()

plt.figure(figsize=[4,6])
sns.scatterplot(x='wt', y='mpg',hue='cyl',data=df)
#plt.title("scatterplot")
plt.xlabel('mpg')
plt.ylabel('cylinder')
plt.show()


plt.figure(figsize=[4,5])
plt.bar(df['mpg'],df['wt'].mean(),color='pink')
plt.xlabel('mpg')
plt.ylabel('wt')
plt.show()
```
